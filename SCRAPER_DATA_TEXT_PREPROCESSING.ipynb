```python
import regex as re

# Fungsi untuk menghapus semua tanda baca dari teks
def remove_punctuation(text):
    # Regular expression untuk mencocokkan semua tanda baca
    pattern = r'\p{P}+'
    # Menghapus tanda baca dari teks
    text = re.sub(pattern, ' ', str(text))
    return text

# Membersihkan tanda baca dari kolom 'Review'
df['cleaning'] = df['Case_folding'].apply(remove_punctuation)
#Mengilangkan semua emoticon
import re

df
```

```python
def remove_all_emoji(text):
    # Menghapus emoji pada kolom review
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002500-\U00002BEF"  # chinese char
                           u"\U00002702-\U000027B0"
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u"\U00010000-\U0010ffff"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u200d"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\ufe0f"  # dingbats
                           u"\u3030"
                           "]+", flags=re.UNICODE)
    # Menghapus emoji dari teks
    return emoji_pattern.sub(r'', text)

# Menghapus emoji dari seluruh kolom 'Review'
df['cleaning'] = df['cleaning'].apply(remove_all_emoji)

# Tampilkan DataFrame yang telah dibersihkan
df
```

2. Normalisasi

```python
import pandas as pd
import re


# Dictionary untuk normalisasi karakter
norm = {
    r'\bgk\b': 'tidak', r'\bga\b': 'tidak', r'\bselma\b': 'selama', r'\bsy\b': 'saya', r'\bgak\b': 'tidak',
    r'\bbner\b': 'benar', r'\bdrmh\b': 'di rumah', r'\bbgt\b': 'banget', r'\bknp\b': 'kenapa',
    r'\bapk\b': 'aplikasi', r'\blelet\b': 'lama', r'\blemot\b': 'lama', r'\blg\b': 'lagi',
    r'\bnmr\b': 'nomor', r'\btolol\b': 'bodoh', r'\byg\b': 'yang', r'\borg\b': 'orang', r'\btp\b': 'tapi',
    r'\baja\b': 'saja', r'\bd\b': 'di', r'\sy\b' : 'saya', r'\bjt\b' :'juta', r'\bbanget\b' : 'bangat'
}

# Fungsi untuk normalisasi teks
def normalisasi(str_text):
    for pattern, replace in norm.items():
        str_text = re.sub(pattern, replace, str_text)
    return str_text

# Menukar nama kolom 'raw_text' menjadi 'cleaning'
df.rename(columns={'raw_text': 'cleaning'}, inplace=True)

# Terapkan fungsi normalisasi pada kolom 'cleaning' dan simpan hasilnya di kolom baru 'Normalization'
df['Normalization'] = df['cleaning'].apply(normalisasi)


# Tampilkan DataFrame hasil
df
```

```python
df = df.drop(columns=['cleaning']) #Menghapus Kolom Cleaning
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Definisikan stop words yang ingin dihilangkan
stop_words = set(stopwords.words('indonesian'))
custom_stop_words = {"kok", "nya", "loh", "gua", "sih", "dong", "bro", "ya", "cie", "lah", "y"}

# Tambahkan custom stop words ke dalam set stop words
stop_words.update(custom_stop_words)

# Contoh data frame 'df' memiliki kolom 'text' yang berisi teks yang ingin dihilangkan stop words-nya
# Anda dapat menyesuaikan dengan nama kolom pada data frame Anda
df['Stopword'] = df['Normalization'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
df
```

```python
# Simpan data frame ke dalam file CSV
df.to_csv('/content/drive/My Drive/datapenelitian/STOPWORD.csv', index=False)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd


# Definisi fungsi stemming
def stemming(text_cleaning):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text_cleaning:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean = " ".join(do)
    return d_clean


# Memanggil fungsi stemming pada kolom 'tokenized'
df['stemmed_Review'] = df['tokenized'].apply(stemming)

# Menyimpan DataFrame yang telah di-stemming ke file CSV
df.to_csv('/content/drive/My Drive/datapenelitian/hasil_stemming.csv', index=False)
```
